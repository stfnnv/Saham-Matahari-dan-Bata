# OHLC Chart Matahari

```
fig, ax1 = plt.subplots(figsize=(15,15))

x = df_mthr['Date']
y1 = df_mthr['Open']
y2 = df_mthr['High']
y3 = df_mthr['Low']
y4 = df_mthr['Close']

ax2 = ax1.twinx()
ax3 = ax1.twinx()
ax4 = ax1.twinx()

ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
ax3.plot(x, y3, 'r-')
ax4.plot(x, y4, 'm-')
```
